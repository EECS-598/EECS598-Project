```python
import os
import xmltodict
import xml.etree.ElementTree as ET
from os import listdir
from os.path import isfile, join
```

```python
tree = ET.parse('metadata/PPMI_3001_sag_3D_FSPGR_BRAVO_straight_S102128_I224574.xml')
root = tree.getroot()
d =dict()
for i in root[0][3][4][6][2]:
    d[list(i.attrib.values())[0]] = i.text
d
```

```python
mypath = 'metadata/'
onlyfiles = [join(mypath,f) for f in listdir(mypath) if isfile(join(mypath, f))]
```

```python
a =[]
for i in onlyfiles:
    tree = ET.parse(i)
    root = tree.getroot()
    
    d =dict()
    for j in root[0][3]:
        if(str.isalnum(j.text)):
#             print(i.tag,i.text)
            d[j.tag] = j.text
    
    d[root[0][3][3][0].tag] = root[0][3][3][0].text
    
    for j in root[0][3][4]:
        if(j.text.strip()!=''):
            d[j.tag] = j.text
            
    for j in root[0][3][4][5]:
        if(j.text.strip()!=''):
            d[j.tag] = j.text
            
    for j in root[0][3][4][6]:
        if(j.text.strip()!=''):
            d[j.tag] = j.text
    
    for j in root[0][3][4][6][2]:
        d[list(j.attrib.values())[0]] = j.text
        
    a.append(d)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(a)
```

```python
df.to_csv('metadata.csv',index=False)
```
