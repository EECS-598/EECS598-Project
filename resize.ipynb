```python
import pandas as pd
import os
import glob
import nibabel as nib
from skimage.transform import resize
```

```python
df = pd.read_csv('metadata_pd_bet.csv');
```

```python
mris = glob.glob('EECS598-Project/PD_BET_2_T1/**/*.nii.gz', recursive=True)
```

```python
count = 1
tot = len(mris)
for mri in mris:
    print((count/tot)*100)
    image = nib.load(mri)
    img  = image.get_data()
    header = image.header
    # image  = resize(img/img.max(), (256, 256,100),order=3)
    uid = int(mri.split('.nii.gz')[0].split('_')[-1][1:])
    
    if (df[(df['imageUID'] == uid)]['Acquisition Plane'] == 'SAGITTAL').iloc[0]:
        scale = (88,256,256)
        header['dim'][1] = 88
        header['dim'][2] = 256
        header['dim'][3] = 256
    else:
        scale = (256,256,88)
        header['dim'][1] = 256
        header['dim'][2] = 256
        header['dim'][3] = 88
    rs_image = resize(img,scale, order=3, mode='reflect', cval=0, clip=True, preserve_range=True, anti_aliasing=True, anti_aliasing_sigma=None)
    



    new_img = nib.Nifti1Image(rs_image, image.affine, header)
    nib.save(new_img, ("EECS598-Project/PD_BET_2_T1_resized/"+ mri.split('/')[-1]))
    count +=1
```
